# Introducción
En este trabajo, diseñaremos un generador de señales para utilizar en las simulaciones que hagamos.

### Señal senoidal
Para esta señal, le pasaremos como argumentos los siguiente valores
- vmax: la amplitud máxima de la senoidal (volts)
- dc: su valor medio (volts)
- ff: la frecuencia (Hz)
- ph: la fase (radianes)
- nn: la cantidad de muestras digitalizada por el ADC (# muestras)
- fs: la frecuencia de muestreo del ADC.

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy import signal

def sin_signal(vmax, dc, ff, ph, nn, fs):
    ts = 1/fs           #Tiempo de muestreo

    #Grilla de sampleo temporal
    #np.linspace devuelve numeros igualmente espaciados en un intervalo
    #flaten Devuelve una copia del array en una dimensión
    tt = np.linspace(0, (nn-1)*ts, nn).flatten()
    
    x = np.array([], dtype=np.float64).reshape(nn,0)
    
    aux = dc + (vmax * np.sin(2*np.pi*ff*tt + ph))
    
    x = np.hstack([x, aux.reshape(nn,1)])
    return x,tt
```

Llamamos a nuestra función y graficamos la señal

```python
x,tt = sin_signal(vmax=1, dc=0, ff=5, ph=0, nn=1000, fs=1000.0)
line_hdls = plt.plot(tt, x)
plt.title('Señal: Senoidal' )
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud [V]')
axes_hdl = plt.gca()    
plt.show()
```

### Señal Cuadrada
Para esta señal, le pasaremos como argumentos los siguiente valores
- vmax: la amplitud máxima de la senoidal (volts)
- dc: su valor medio (volts)
- ff: la frecuencia (Hz)
- duty: duty de la señal
- nn: la cantidad de muestras digitalizada por el ADC (# muestras)
- fs: la frecuencia de muestreo del ADC.

```python
def square_signal(vmax, dc, ff, duty, nn, fs):
    ts = 1/fs           #Tiempo de muestreo

    #Grilla de sampleo temporal
    #np.linspace devuelve numeros igualmente espaciados en un intervalo
    #flaten Devuelve una copia del array en una dimensión
    tt = np.linspace(0, (nn-1)*ts, nn).flatten()
    
    x = np.array([], dtype=np.float64).reshape(nn,0)
    
    aux = dc + (vmax * signal.square(2 * np.pi* ff *tt, duty));
    
    x = np.hstack([x, aux.reshape(nn,1)])
    return x,tt
```

Llamamos a nuestra función y la señal resulta

```python
x,tt = square_signal(vmax=1, dc=0, ff=5, duty=0.3, nn=1000, fs=1000.0)

plt.figure(1)
line_hdls = plt.plot(tt, x)
plt.title('Señal: Cuadrada' )
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud [V]')
axes_hdl = plt.gca()    
plt.show()
```

### Señal diente de sierra
Para esta señal, le pasaremos como argumentos los siguiente valores
- vmax: la amplitud máxima de la senoidal (volts)
- dc: su valor medio (volts)
- ff: la frecuencia (Hz)
- nn: la cantidad de muestras digitalizada por el ADC (# muestras)
- fs: la frecuencia de muestreo del ADC.

```python
def sawtooth_signal(vmax, dc, ff, nn, fs):
    ts = 1/fs           #Tiempo de muestreo

    #Grilla de sampleo temporal
    #np.linspace devuelve numeros igualmente espaciados en un intervalo
    #flaten Devuelve una copia del array en una dimensión
    tt = np.linspace(0, (nn-1)*ts, nn).flatten()
    
    x = np.array([], dtype=np.float64).reshape(nn,0)
    
    aux = dc + (vmax * signal.sawtooth(2 * np.pi* ff *tt));
    
    x = np.hstack([x, aux.reshape(nn,1)])
    return x,tt
```

Llamamos a nuestra función y la señal resulta

```python
x,tt = sawtooth_signal(vmax=1, dc=0, ff=5, nn=1000, fs=1000.0)

plt.figure(1)
line_hdls = plt.plot(tt, x)
plt.title('Señal: Diente de sierra' )
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud [V]')
axes_hdl = plt.gca()    
plt.show()

    
```
