# Introducción
En este trabajo, simularemos el comportamiento de un ADC y cómo sus parámetros, tamaño de palabra de B bits y frecuencia de muestreo fs afectan a una señal (aproximadamente) analógica.

### Señal senoidal
La señal senoidal tendra las siguientes características
- vmax = 1V
- dc = 0V
- ff = 1Hz
- ph = 0
- nn = 1000
- fs = 1000.

```python
import genSign as signal_gen
import numpy as np
import matplotlib.pyplot as plt

N = 1000
fs = 1000

x,tt = signal_gen.sin_signal(vmax = 1, dc = 0, ff = 1, ph = 0, nn = N, fs = fs)
```

Graficamos la señal senoidal y nos queda

```python
line_hdls = plt.plot(tt, x)
plt.title('Señal: Senoidal' )
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud [V]')
axes_hdl = plt.gca()    
plt.show()
```

A continuación, el código de la función encargada de simular nuestro ADC donde
- Vf : Rango de tensión del ADC 
- B : Bits del ADC

```python
def quantifySignal(Vf, B):
    y = np.full(
    shape = N,
    fill_value = 0,
    dtype= float)
    
    q = Vf/(2**(B-1))
    for k in range(N-1):
        y[k] = np.round(x[k]/q) * q
    return y
```

Si cuantificamos la señal utilizando un rango de tensión +-2V y 4 bits

```python
plt.figure(1)
line_hdls = plt.plot(tt, x)
axes_hdl = plt.gca()        
line_hdls = plt.plot(tt, quantifySignal(2, 4))
axes_hdl = plt.gca()    
plt.xlabel("Tiempo (s)")
plt.ylabel("Tensión (V)")
plt.legend(["Señal entrada", "Señal muestreada por el ADC"])
plt.show()
```

Utilizando un ADC de 8 bits, obtenemos

```python
plt.figure(1)
line_hdls = plt.plot(tt, x)
axes_hdl = plt.gca()        
line_hdls = plt.plot(tt, quantifySignal(2, 8))
axes_hdl = plt.gca()    
plt.xlabel("Tiempo (s)")
plt.ylabel("Tensión (V)")
plt.legend(["Señal entrada", "Señal muestreada por el ADC"])
plt.show()
```
