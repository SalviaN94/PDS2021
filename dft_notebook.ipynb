### Código

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy import signal

def custom_dft(x, N):
    y = np.full(
    shape=N, 
    fill_value=0+0j,
    dtype = "complex_")
    
    for k in range(N-1):
        for n in range(N):
            y[k] += x[n] * np.exp(-2j * np.pi *(k * n)/N)
    return y
```

Llamando a al función y graficando magnitud y fase de la transformada

```python
import genSign as signal_gen

N = 1000
fs = 1000
ts = 1/fs # tiempo de muestreo
df = fs/N # resolución espectral

# grilla de sampleo frecuencial
ff = np.linspace(0, (N-1), N) * df
    
x,tt = signal_gen.sin_signal(vmax=1, dc=0, ff=1, ph=0, nn = N, fs = fs)
y = custom_dft(x, N)

plt.figure(1)
line_hdls = plt.plot(tt, x)
axes_hdl = plt.gca()        
plt.figure(2)
plt.xlim([0, 10])
line_hdls = plt.stem(ff, np.absolute(y))
axes_hdl = plt.gca()    
plt.figure(3)
plt.xlim([0, 10])
line_hdls = plt.stem(ff, np.angle(y))
axes_hdl = plt.gca();
plt.show()
```
